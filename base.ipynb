```python
import pandas as pd
```

```python
df = pd.read_csv(r"brasileirao.csv")
df
```

```python
# Criando novas estatiscas com dados ja existentes 
df["percentual_derrotas"] = df["loss"] / df["played"] * 100
df["media_gols_sofridos"] = df["goals_taken"] / df["played"] 
df["media_gols_marcados"] = df["goals"] / df["played"]
df
```

```python
filtro_lanterna = df["place"] == 20
df_rebaixados = df[filtro_lanterna]
df_rebaixados
```

TIMES QUE MAIS VEZES FORAM LANTERNAS DA COMPETIÇÃO


```python
df_rebaixados.groupby("team")["place"].count().sort_values(ascending=False)
```

Comparação de 16º com 17º


```python

decimosexto = df[df["place"] == 16]
rebaixados = df[df["place"] == 17]

print("media de gols marcados por 16º: ", decimosexto["goals"].mean())
print("media de gols marcados por 17º: ",  rebaixados["goals"].mean())
print("-" * 50)
print("media de gols sofridos por 16º: ", decimosexto["goals_taken"].mean())
print("media de gols sofridos por 17º: ", rebaixados["goals_taken"].mean())
print("-" * 50)
print("media de vitorias por 16º: ", decimosexto["won"].mean())
print("media de vitorias por 17º: ", rebaixados["won"].mean())
print("-" * 50)
print("Percentual de derrotas 16º:", decimosexto["percentual_derrotas"].mean())
print("Percentual de derrotas 17º:", rebaixados["percentual_derrotas"].mean())
print("-" * 50)
print("media de pontos por 16º: ", decimosexto["points"].mean())
print("media de pontos por 17º: ", rebaixados["points"].mean())
print("-" * 50)    
```
